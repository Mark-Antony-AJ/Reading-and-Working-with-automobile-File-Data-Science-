```python
auto_temp1=auto[["make","price"]]
group_auto=auto_temp1.groupby(["make"]).mean().sort_values(ascending=False,by="price")
group_auto
```

```python
temp_df=auto[["make","price"]].groupby(["make"])
temp_df.head()
```
